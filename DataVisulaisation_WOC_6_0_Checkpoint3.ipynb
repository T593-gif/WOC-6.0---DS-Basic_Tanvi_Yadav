```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'/content/CarPrice_Assignment.csv')
dataset
```

```python
sns.scatterplot(x="car_ID",y="fueltype",data=dataset)
plt.show()
```

```python
sns.lineplot(x="car_ID",y="price",data=dataset,hue="fueltype")
plt.show()
```

```python
sns.lineplot(x="car_ID",y="horsepower",data=dataset,hue="fueltype")
plt.show()
```

```python
sns.scatterplot(x="car_ID",y="enginesize",data=dataset,hue="fueltype")
plt.show()
```

```python
sns.scatterplot(x="car_ID",y="carbody",data=dataset,hue="fueltype")
plt.show()
```

```python
missing_values = data_frame.isnull().sum()
print(missing_values)
```

```python
data_frame = data_frame.dropna()
print(data_frame)
```

```python
columns_to_drop = ['Duration in Seconds', 'Hashtags', 'Date of the Last Comment', 'Intern Who Collected the Data']
data_frame = data_frame.drop(columns = columns_to_drop)
print(data_frame)
```

```python
data_frame.head(5)
```

```python

categorical_columns = ['Creator Gender', 'Premiered or Not', 'Language of the Video', 'Subtitle', 'Video Description', 'Video Title', 'Creator Name']
data_frame_encoded = pd.get_dummies(data_frame, columns=categorical_columns)

print(data_frame_encoded)
```

```python

numerical_columns = ['Video Views', 'Total Channel Subcribers', 'Total Chanel Views', 'No of Comments', 'No of Videos the Channel']

data_frame[numerical_columns] = data_frame[numerical_columns].apply(lambda x: x.str.replace(',', '').astype(float), axis=1)

# Display summary statistics
summary_statistics = data_frame[numerical_columns].describe()
print(summary_statistics)
```

```python

def remove_outliers_iqr(column):
    Q1 = data_frame[column].quantile(0.25)
    Q3 = data_frame[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data_frame[(data_frame[column] >= lower_bound) & (data_frame[column] <= upper_bound)]

# Remove outliers for each numerical column
for column in numerical_columns:
    data_frame = remove_outliers_iqr(column)

# Display the cleaned DataFrame without outliers
print(data_frame)
```
